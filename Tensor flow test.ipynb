```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.standard_normal((200, 1))
```

```python
# parametro w=2 e d=3
y = 2 * X + 3
```

```python
plt.plot(X, y)
plt.grid(True)
```

```python
#creazione del neurone in questo caso ce n'è uno
model = tf.keras.Sequential([tf.keras.layers.Dense(1, input_shape=[1])])
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
model.compile(
    loss=tf.keras.losses.mean_squared_error, optimizer=tf.keras.optimizers.Adam()
)
```

```python
history = model.fit(x=X, y=y, epochs=1000)
```

```python
y_pred = model.predict(X)
```

```python
np.c_[y, y_pred]
```

```python
#vediamo l'errore medio
tf.keras.losses.mean_squared_error(y[:, 0], y_pred[:, 0])
```

```python
loss_hist = history.history["loss"]
plt.plot(range(1000), loss_hist)
plt.grid(True)
```
